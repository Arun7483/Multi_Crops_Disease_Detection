```python
import os
import cv2
import yaml
import random
import shutil
import numpy as np
from glob import glob
from tqdm import tqdm
from matplotlib import pyplot as plt
from collections import defaultdict
from albumentations import (
    HorizontalFlip, VerticalFlip, Rotate,
    RandomBrightnessContrast, Compose, BboxParams,
    ShiftScaleRotate, Affine
)

class EnhancedDataBalancer:
    def __init__(self, root, data_types, output_dir):
        self.root = root
        self.data_types = data_types
        self.output_dir = output_dir
        self.class_dict = {}
        self.analysis_data = defaultdict(dict)
        self._verify_and_create_paths()
        self._copy_config_file()
        self._load_class_names()
        self._copy_original_data()
        self._process_datasets()

    def _verify_and_create_paths(self):
        os.makedirs(self.output_dir, exist_ok=True)
        for dtype in self.data_types:
            os.makedirs(f"{self.output_dir}/{dtype}/images", exist_ok=True)
            os.makedirs(f"{self.output_dir}/{dtype}/labels", exist_ok=True)
            if not os.path.exists(f"{self.root}/{dtype}"):
                os.makedirs(f"{self.root}/{dtype}/images", exist_ok=True)
                os.makedirs(f"{self.root}/{dtype}/labels", exist_ok=True)

    def _copy_config_file(self):
        src = f"{self.root}/data.yaml"
        dst = f"{self.output_dir}/data.yaml"
        if os.path.exists(src):
            shutil.copyfile(src, dst)
            print(f"‚úÖ Copied config file to {dst}")
        else:
            raise FileNotFoundError(f"Missing config file: {src}")

    def _load_class_names(self):
        config_path = f"{self.root}/data.yaml"
        with open(config_path) as f:
             data = yaml.safe_load(f)
        self.class_dict = {i: name for i, name in enumerate(data['names'])}

    def _copy_original_data(self):
        for dtype in self.data_types:
            for folder in ['images', 'labels']:
                src = f"{self.root}/{dtype}/{folder}"
                dst = f"{self.output_dir}/{dtype}/{folder}"
                if os.path.exists(src):
                    shutil.copytree(src, dst, dirs_exist_ok=True)

    def _process_datasets(self):
        for dtype in self.data_types:
            class_counter = defaultdict(int)
            image_dir = f"{self.output_dir}/{dtype}/images"
            
            for img_path in tqdm(glob(f"{image_dir}/*"), desc=f"Processing {dtype}"):
                label_path = img_path.replace("/images/", "/labels/").rsplit('.', 1)[0] + '.txt'
                
                if not os.path.exists(label_path):
                    continue
                    with open(label_path) as f:
                    lines = [line.strip() for line in f.readlines() if line.strip()]
                
                present_classes = set()
                for line in lines:
                    try:
                        class_id = int(float(line.split()[0]))
                        present_classes.add(self.class_dict[class_id])
                    except (ValueError, IndexError):
                        continue
                
                for cls in present_classes:
                    class_counter[cls] += 1
            
            self.analysis_data[dtype] = dict(class_counter)

    def analyze_balance(self):
        for dtype in self.data_types:
            total = sum(self.analysis_data[dtype].values())
            num_classes = len(self.class_dict)
            avg = total / num_classes
             print(f"\nüîç {dtype.upper()} ANALYSIS")
            print(f"üìä Total Images: {total}")
            print(f"‚öñÔ∏è Average per Class: {avg:.1f}")
            print("\nüìà Class Distribution:")
            for cls, count in self.analysis_data[dtype].items():
                diff = count - avg
                status = "‚¨áÔ∏è UNDER" if diff < -avg*0.2 else "‚¨ÜÔ∏è OVER" if diff > avg*0.2 else "‚úÖ BALANCED"
                print(f"- {cls:15}: {count:4} ({status})")
            
            self._plot_distribution(dtype, "Original Distribution")

    def balance_strategy(self):
        self.augmentor = AdvancedAugmentor(
            self.root, 
            self.output_dir,
            self.class_dict
        )
        for dtype in self.data_types:
            image_dir = f"{self.output_dir}/{dtype}/images"
            total = len(glob(f"{image_dir}/*"))
            avg = total / len(self.class_dict)
            
            print(f"\n‚öôÔ∏è Balancing {dtype} dataset...")
            
            # Ch·ªâ augment c√°c l·ªõp d∆∞·ªõi m·ª©c trung b√¨nh
            target_counts = {
                cls: int(avg) 
                for cls, count in self.analysis_data[dtype].items() 
                if count < avg
            }
            
            self.augmentor.augment_dataset(dtype, target_counts, self.analysis_data[dtype])
            
            # C·∫≠p nh·∫≠t s·ªë l∆∞·ª£ng sau augment
            new_counts = defaultdict(int)
            new_images = glob(f"{self.output_dir}/{dtype}/images/*")
            for img_path in new_images:
                label_path = img_path.replace("images", "labels").replace(".jpg", ".txt")
                                try:
                    with open(label_path) as f:
                        lines = [line.strip() for line in f.readlines() if line.strip()]
                    for line in lines:
                        parts = line.split()
                        if len(parts) != 5:
                            continue
                        class_id = int(float(parts[0]))
                        new_counts[self.class_dict[class_id]] += 1
                except Exception as e:
                    print(f"Error processing {label_path}: {str(e)}")

    def _plot_distribution(self, dtype, title):
        classes = sorted(self.analysis_data[dtype].keys())
        counts = [self.analysis_data[dtype][cls] for cls in classes]
        
        plt.figure(figsize=(15, 8))
        bars = plt.bar(classes, counts, color='#2ecc71', edgecolor='black')
        plt.title(f"{dtype.upper()} - {title}", fontsize=14, pad=20)
        plt.xlabel("Class Names", fontsize=12, labelpad=10)
        plt.ylabel("Image Count", fontsize=12, labelpad=10)
        plt.xticks(rotation=45, ha='right', fontsize=10)
        plt.yticks(fontsize=10)
        
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height,
                     f'{int(height)}',
                     ha='center', va='bottom', fontsize=9)
        
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()
     def print_final_counts(self):
        print("\nüìä FINAL DATASET COUNTS")
        for dtype in self.data_types:
            original_count = len(glob(f"{self.root}/{dtype}/images/*"))
            new_count = len(glob(f"{self.output_dir}/{dtype}/images/*"))
            print(f"‚úÖ {dtype.upper()}:")
            print(f"   Original: {original_count} images")
            print(f"   Augmented: {new_count } images")
            print(f"   Total increase: {new_count - original_count} (+{(new_count / original_count - 1) * 100:.1f}%)")

    def analyze_balanced_data(self):
        for dtype in self.data_types:
            balanced_image_dir = f"{self.output_dir}/{dtype}/images"
            balanced_images = glob(f"{balanced_image_dir}/*")
            
            class_counter = defaultdict(int)
            
            for img_path in tqdm(balanced_images, desc=f"Analyzing balanced {dtype}"):
                label_path = img_path.replace("/images/", "/labels/").replace(".jpg", ".txt")
                
                if not os.path.exists(label_path):
                    continue
                     try:
                    with open(label_path, 'r') as f:
                        lines = [line.strip() for line in f.readlines() if line.strip()]
                except Exception as e:
                    print(f"Error reading {label_path}: {str(e)}")
                    continue
                    
                present_classes = set()
                for line in lines:
                    parts = line.split()
                    if len(parts) != 5:
                        continue
                     class_id = int(float(parts[0]))
                    present_classes.add(self.class_dict.get(class_id, "unknown"))
                        
                for cls in present_classes:
                    class_counter[cls] += 1
                    
            total = sum(class_counter.values())
            print(f"\nüìä {dtype.upper()} BALANCED ANALYSIS")
            print(f"üìà Total Images: {total}")
            print("üìä Class Distribution:")
            for cls, count in sorted(class_counter.items(), key=lambda x: x[1], reverse=True):
                percentage = (count / total) * 100 if total > 0 else 0
                print(f"- {cls:25}: {count:4} ({percentage:.2f}%)")
            
            self._plot_balanced_distribution(dtype, class_counter)
    
    def _plot_balanced_distribution(self, dtype, class_counter):
        classes = sorted(class_counter.keys(), key=lambda x: class_counter[x], reverse=True)
        counts = [class_counter[cls] for cls in classes]
        
        plt.figure(figsize=(18, 8))
         bars = plt.bar(classes, counts, color='#3498db', edgecolor='black')
        plt.title(f"{dtype.upper()} - Balanced Distribution", fontsize=14, pad=20)
        plt.xlabel("Class Names", fontsize=12, labelpad=10)
        plt.ylabel("Image Count", fontsize=12, labelpad=10)
        plt.xticks(rotation=45, ha='right', fontsize=10)
        plt.yticks(fontsize=10)
        
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height,
                     f'{int(height)}',
                     ha='center', va='bottom', fontsize=9)
        
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()
class AdvancedAugmentor:
    def __init__(self, root, output_dir, class_dict):
        self.root = root
        self.output_dir = output_dir
        self.class_dict = class_dict
        self._init_transforms()

    def _init_transforms(self):
        self.strong_aug = Compose([
            HorizontalFlip(p=0.5),
            VerticalFlip(p=0.3),
            ShiftScaleRotate(shift_limit=0.1, scale_limit=0.2, rotate_limit=30, p=0.8),
            RandomBrightnessContrast(p=0.5),
            Rotate(limit=20, p=0.4),
            Affine(scale=0.3, p=0.4)
        ], bbox_params=BboxParams(
            format='yolo',
            min_visibility=0.4,
            label_fields=['class_labels']
        ))

        self.mild_aug = Compose([
            HorizontalFlip(p=0.3),
            RandomBrightnessContrast(p=0.2)
        ], bbox_params=BboxParams(
            format='yolo',
            label_fields=['class_labels']
        ))

    def augment_dataset(self, dtype, target_counts, current_counts):
        original_images = glob(f"{self.root}/{dtype}/images/*")
        
        for class_name, target_count in target_counts.items():
            current_count = current_counts.get(class_name, 0)
            if current_count >= target_count:
                continue
            
            needed = target_count - current_count
             class_images = [img for img in original_images if self._has_class(img, class_name)]
            
            if not class_images:
                print(f"‚ö†Ô∏è No images found for class {class_name}, skipping...")
                continue
            
            print(f"üîÑ Augmenting {class_name}: {current_count} -> {target_count}")
            for _ in tqdm(range(needed), desc=class_name):
                img_path = random.choice(class_images)
                label_path = img_path.replace("images", "labels").replace(".jpg", ".txt")
                
                image = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
                with open(label_path) as f:
                    lines = [line.strip() for line in f.readlines() if line.strip()]
                
                bboxes, class_ids = [], []
                for line in lines:
                    parts = list(map(float, line.split()))
                    class_ids.append(int(parts[0]))
                    bboxes.append(parts[1:5])
                      
                try:
                    if needed > 100:
                        transformed = self.strong_aug(image=image, bboxes=bboxes, class_labels=class_ids)
                    else:
                        transformed = self.mild_aug(image=image, bboxes=bboxes, class_labels=class_ids)
                except Exception as e:
                    print(f"Error processing {img_path}: {str(e)}")
                    continue
                
                self._save_data(transformed['image'], transformed['bboxes'], transformed['class_labels'], img_path, dtype)

    def _has_class(self, img_path, target_class):
         label_path = img_path.replace("images", "labels").replace(".jpg", ".txt")
        try:
            with open(label_path) as f:
                lines = [line.strip() for line in f.readlines() if line.strip()]
            for line in lines:
                class_id = int(line.split()[0])
                if self.class_dict.get(class_id, "unknown") == target_class:
                    return True
            return False
        except Exception as e:
            print(f"Error checking class in {img_path}: {str(e)}")
            return False

    def _save_data(self, image, bboxes, class_ids, original_path, dtype):
        base_name = os.path.basename(original_path)
        unique_id = random.randint(100000, 999999)
        new_name = f" aug_{unique_id}_{base_name}"
        
        output_image_path = os.path.join(self.output_dir, dtype, "images", new_name)
        output_label_path = os.path.join(self.output_dir, dtype, "labels", new_name.replace(".jpg", ".txt"))
        cv2.imwrite(output_image_path, cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
        
        with open(output_label_path, 'w') as f:
            for cls_id, bbox in zip(class_ids, bboxes):
                if all(0 <= x <= 1 for x in bbox):
                    f.write(f"{int(cls_id)} {' '.join(f'{x:.6f}' for x in bbox)}\n")


class EnhancedVisualizer:
    def __init__(self, dataset_dir, data_types, class_dict):
        self.dataset_dir = dataset_dir
        self.data_types = data_types
        self.class_dict = class_dict

    def visualize_samples(self, with_boxes=True, samples_per_set=4):
        for dtype in self.data_types:
            image_dir = f"{self.dataset_dir}/{dtype}/images"
            label_dir = f"{self.dataset_dir}/{dtype}/labels"
            image_paths = glob(f"{image_dir}/*.jpg")
            if not image_paths:
                print(f"‚ö†Ô∏è No images found in {image_dir}")
                continue

            sample_images = random.sample(image_paths, min(samples_per_set, len(image_paths)))

            fig, axes = plt.subplots(1, len(sample_images), figsize=(20, 5))
            if len(sample_images) == 1:
                axes = [axes]

            for ax, img_path in zip(axes, sample_images):
                img = cv2.imread(img_path)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                
                label_path = img_path.replace("images", "labels").replace(".jpg", ".txt")

                if with_boxes and os.path.exists(label_path):
                    with open(label_path, "r") as f:
                        lines = f.readlines()
                     h, w, _ = img.shape
                    for line in lines:
                        parts = line.strip().split()
                        if len(parts) != 5:
                            continue
                        
                        class_id = int(parts[0])
                        x_center, y_center, box_w, box_h = map(float, parts[1:])
                        x1 = int((x_center - box_w / 2) * w)
                        y1 = int((y_center - box_h / 2) * h)
                        x2 = int((x_center + box_w / 2) * w)
                        y2 = int((y_center + box_h / 2) * h)

                        label = self.class_dict.get(class_id, "Unknown")
                        color = (0, 255, 0)
                        cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
                        cv2.putText(img, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)
                        ax.imshow(img)
                ax.axis("off")
                ax.set_title(f"{dtype.upper()} Sample")

            plt.show()


if __name__ == "__main__":
    root_path = "C:\Users"
    output_dir = "/kaggle/working/balanced_dataset"
    data_types = ["train", "valid", "test"]
    
    balancer = EnhancedDataBalancer(root_path, data_types, output_dir)
    balancer.analyze_balance()
    balancer.balance_strategy()  # Ch·ªâ augment c√°c l·ªõp d∆∞·ªõi m·ª©c trung b√¨nh
    balancer.analyze_balanced_data()
    balancer.print_final_counts()
    
    visualizer = EnhancedVisualizer(output_dir, data_types, balancer.class_dict)
    visualizer.visualize_samples(with_boxes=True, samples_per_set=4)
```

```python
import os
import cv2
import yaml
import random
import numpy as np
from PIL import Image
from glob import glob
from matplotlib import pyplot as plt
from torchvision import transforms as T

class Visualization:
    def __init__(self, root, data_types, n_ims, rows, cmap=None):
        self.root = root
        self.n_ims, self.rows = n_ims, rows
        self.cmap, self.data_types = cmap, data_types
        self.colors = ["firebrick", "darkorange", "blueviolet"]

        self.get_cls_names()
        self.get_bboxes()

    def get_cls_names(self):
        yaml_path = os.path.join(self.root, "data.yaml")
        print(f"Loading class names from: {yaml_path}")  # Debugging line

        if not os.path.exists(yaml_path):
            raise FileNotFoundError(f"data.yaml not found at {yaml_path}")

        with open(yaml_path, 'r') as file:
            data = yaml.safe_load(file)

        self.class_dict = {index: name for index, name in enumerate(data['names'])}
        print(f"Loaded class names: {self.class_dict}")  # Debugging line

    def get_bboxes(self):
        self.vis_datas, self.analysis_datas, self.im_paths = {}, {}, {}

        for data_type in self.data_types:
            all_bboxes, all_analysis_datas = [], {}
            im_paths = glob(os.path.join(self.root, data_type, "images", "*"))

            for im_path in im_paths:
                bboxes = []
                im_ext = os.path.splitext(im_path)[-1]
                lbl_path = im_path.replace(im_ext, ".txt").replace(f"{data_type}/images", f"{data_type}/labels")

                if not os.path.isfile(lbl_path):
                    print(f"Warning: No label file for {im_path}")
                    all_bboxes.append([])  # Ensure index alignment
                    continue

                with open(lbl_path) as f:
                    meta_data = f.readlines()

                for data in meta_data:
                    parts = data.strip().split()[:5]
                    cls_name = self.class_dict[int(parts[0])]
                    bboxes.append([cls_name] + [float(x) for x in parts[1:]])

                    all_analysis_datas[cls_name] = all_analysis_datas.get(cls_name, 0) + 1

                all_bboxes.append(bboxes)

            self.vis_datas[data_type] = all_bboxes
            self.analysis_datas[data_type] = all_analysis_datas
            self.im_paths[data_type] = im_paths

    def plot(self, rows, cols, count, im_path, bboxes):
        if count > rows * cols:  # Ensure count doesn't exceed the subplot grid
            return count

        plt.subplot(rows, cols, count)
        or_im = np.array(Image.open(im_path).convert("RGB"))
        height, width, _ = or_im.shape

        for bbox in bboxes:
            if len(bbox) < 5:
                continue  # Skip malformed labels

            _, x_center, y_center, w, h = bbox

            # Convert YOLO format to pixel values
            x_min = int((x_center - w / 2) * width)
            y_min = int((y_center - h / 2) * height)
            x_max = int((x_center + w / 2) * width)
            y_max = int((y_center + h / 2) * height)

            color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
            cv2.rectangle(img=or_im, pt1=(x_min, y_min), pt2=(x_max, y_max), color=color, thickness=3)

        plt.imshow(or_im)
        plt.axis("off")
        plt.title(f"There is (are) {len(bboxes)} object(s) in the image.")

        return count + 1

    def data_analysis(self, save_name, color):
        print(f"Performing data analysis for {save_name}...\n")

        width, text_width, text_height = 0.7, 0.05, 2
        cls_names = list(self.analysis_datas[save_name].keys())
        counts = list(self.analysis_datas[save_name].values())

        _, ax = plt.subplots(figsize=(15, 6))
        indices = np.arange(len(counts))

        ax.bar(indices, counts, width, color=color)
        ax.set_xlabel("Class Names", color="black")
        ax.set_xticklabels(cls_names, rotation=90)
        ax.set_xticks(indices)
        ax.set_ylabel("Data Counts", color="black")
        ax.set_title(f"{save_name.upper()} Dataset Class Distribution")

        for i, v in enumerate(counts):
            ax.text(i - text_width, v + text_height, str(v), color="royalblue")

    def visualization(self):
        fig = plt.figure(figsize=(15, 10))  # Set figure size
        count = 1
        cols = 4  # Number of columns for visualization
        max_images = self.rows * cols  # Ensure we do not exceed subplot grid

        for save_name in self.data_types:
            images = self.im_paths.get(save_name, [])[:self.n_ims]  # Get available images
            bboxes_list = self.vis_datas.get(save_name, [])  # Get bounding boxes safely

            for i, im_path in enumerate(images):
                if i >= len(bboxes_list):  # Prevent IndexError
                    print(f"Skipping {im_path} (No bounding box data)")
                    continue

                if count > max_images:  # Ensure count does not exceed grid
                    break

                count = self.plot(self.rows, cols, count, im_path, bboxes_list[i])

        plt.tight_layout()
        plt.show()

    def analysis(self):
        for save_name, color in zip(self.data_types, self.colors):
            self.data_analysis(save_name, color)


# Corrected root path (pointing to the dataset directory, not data.yaml)
root = r"C:\Users\krypt\.cache\kagglehub\datasets\andresmgs\plantdec\versions\7"

# Create instance and run analysis & visualization
vis = Visualization(root=root, data_types=["train", "valid", "test"], n_ims=20, rows=5, cmap="rgb")
vis.analysis()
vis.visualization()  # Now calling visualization properly
```
